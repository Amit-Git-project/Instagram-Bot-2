### Problem 1 : Login to your Instagram
Login to your Instagram Handle  
Submit with sample username and password

```python
def login(username,password):
    try:
        #locating username box and assigning username
        username_box = driver.find_element("name","username")
        username_box.send_keys(username)
        
        # locating password box and assigning password
        password_box = driver.find_element("name","password")
        password_box.send_keys(password)

        # locating login button and clicking it 
        login_btn = wait.until(EC.visibility_of_element_located
                               (("xpath","//button[contains(@class,' _acan _acap _acas _aj1- _ap30')]")))
        login_btn.click()
        
        # Handling "Save login info" by clicking not now
        not_now_element = wait.until(EC.visibility_of_element_located(
            ("class name","_ac8f")))
        not_now_element.click()
        
        # Handling "Turn on notifications" by clicking not now
        not_now_element = wait.until(EC.visibility_of_element_located(
            ("xpath","//button[contains(@class,'_a9-- _ap36 _a9_1')]")))
        not_now_element.click()
        
    except:
        
        # Handling any encountered errors during this process
        print("Some Error Encountered")
```

### Problem 2 : Type for “food” in search bar
Type for “food” in search bar and print all the names of the Instagram Handles that are displayed in list after typing “food”  
Note : Make sure to avoid printing hashtags  

```python
def Search(text):
    try:
        # locating Search icon
        search_icon =  wait.until(
                EC.presence_of_element_located(("xpath","/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[1]/div/div/div/div/div[2]/div[2]/span/div/a/div")
                                                   ))
        
        # clicking it to appear search icon
        search_icon.click()
        
        # Waiting and Locating Search bar to write text
        search_bar = wait.until(EC.presence_of_element_located(("xpath","//input[@aria-label='Search input']")))
        
        #Sending text "food" in Search Bar
        search_bar.send_keys(text)
        
        # waiting for dropdown menu appear
        time.sleep(5)
        
        dropdownMenu = wait.until(EC.presence_of_element_located(("xpath","//div[contains(@class,'x6s0dn4 x78zum5 xdt5ytf x5yr21d x1odjw0f x1n2onr6 xh8yej3')]")
                                  ))
        # Waiting for all instagram handle to appear
        instagram_handles = WebDriverWait(dropdownMenu,10).until(EC.presence_of_all_elements_located(
            ("xpath",".//span[contains(@class,'x1lliihq x1plvlek xryxfnj x1n2onr6 x193iq5w xeuugli x1fj9vlw x13faqbe x1vvkbs x1s928wv xhkezso x1gmr53x x1cpjm7i x1fgarty x1943h6x x1i0vuye xvs91rp x1s688f x5n08af x10wh9bi x1wdrske x8viiok x18hxmgj')]"))
          )
        
        # Collecting all instagram handle names excluding "#" tag
        names = [i.text for i in instagram_handles if len(i.text)>0 and i.text[0] != "#"]
        
        #closing search icon
        search_icon.click()
        
        # returning the list of instagram handles
        return names
        
    
    except:
        
        # Handling any encountered errors during this process
        print("Some Error Encountered")
```

```python
names = Search("food")
for name in names:
    print(name)
```

### Problem 3 : Searching and Opening a profile
Searching and Opening a profile using   
Open profile of “So Delhi” 

```python
def SearchOpenProfile(text):
    try:
        #Locating search icon
        search_icon =  wait.until(
                EC.presence_of_element_located(("xpath","/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[1]/div/div/div/div/div[2]/div[2]/span/div/a/div")
                                                   ))
        
        # clicking it to appear search icon
        search_icon.click()
        
        #Opening Search Bar
        search_bar = wait.until(EC.presence_of_element_located(("xpath","//input[@aria-label='Search input']")))
        
        search_bar.clear()
        # Sending text "So Delhi" to search bar
        search_bar.send_keys(text)
        
        #Waiting for dialogue Box to open up
        time.sleep(5)
        
        # fetching "So Delhi" element
        profile_element = wait.until(EC.presence_of_element_located(
        ("xpath","//div[contains(@class,'x9f619 x1n2onr6 x1ja2u2z x78zum5 xdt5ytf x2lah0s x193iq5w xeuugli x1iyjqo2')]"))
                  )
        
        # Clicking "So Delhi"
        profile_element.click()
        
        # Waiting for profile to open up properly
        time.sleep(3)
        
        #closing search icon
        search_icon.click()
    except:
        
        # Handling any encountered errors during this process
        print("Some Error Encountered")
```

### Problem 4 : Follow/Unfollow given handle
Follow/Unfollow given handle -   
1.Open the Instagram Handle of “So Delhi”  
2.Start following it. Print a message if you are already following  
3.After following, unfollow the instagram handle. Print a message if you have already unfollowed.

```python
def follow():
    try:
         # Waiting for the presence of the follow button
        follow_btn = wait.until(EC.presence_of_element_located(
        ("xpath","//button[contains(@class,' _acan _acap _acas _aj1- _ap30')]")))
        
        # Checking if the button text is 'Follow'
        if follow_btn.text == "Follow":
            
            # Clicking the follow button if it says 'Follow'
            follow_btn.click()
        else:
            
            # If the button text is not 'Follow', indicating already following
            print("Already Following")
    except:
        
        # Handling any encountered errors during this process
        print("Some Error Encountered")
```

```python
def unfollow():
    try:
        # Waiting for the presence of the following button
        following_btn = wait.until(EC.presence_of_element_located(
        ("xpath","//button[contains(@class,' _acan _acap _acat _aj1- _ap30')]")))
        
        # Checking if the button text is 'Following'
        if following_btn.text == 'Following':
            
            # Clicking the following button if it says 'Following'
            following_btn.click()
            
            # Waiting for the presence of the unfollowing button and choosing it from the DropDown menu
            unfollow_btn = wait.until(EC.presence_of_all_elements_located(
                ("xpath","//div[contains(@class,'x9f619 x1n2onr6 x1ja2u2z x78zum5 xdt5ytf x2lah0s x193iq5w xeuugli x1iyjqo2')]"))
                                     )[-1]
            # Clicking the unfollow button
            unfollow_btn.click()

        else:
            print("Already Unfollowed")
    except:
        
        # Handling any encountered errors during this process
        print("Some Error Encountered")
```

```python
follow()
unfollow()
```

### Problem 5 : Like/Unlike posts
Like/Unlike posts  
1.Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.  
2.Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python
# Opening profile of "dilsefoodie"
SearchOpenProfile("dilsefoodie")
```

```python
def fetchingPosts(NumberofPosts):
    
    try:
        # Waiting for the profile to properly appear
        time.sleep(1.5)
        
        #Calculating currernt height of the window
        curr_height = driver.execute_script("return document.body.scrollHeight")

        # Intialising a while loop with "True condition"
        while True:

            # Scrolling window till the curr_height value
            driver.execute_script(f"window.scrollTo(0,{curr_height})")

            # Waiting for new elements to appear
            time.sleep(3)

            # Calculating new height after scroll
            new_height = driver.execute_script("return document.body.scrollHeight")

            # Fetching Posts 
            post_elements = wait.until(EC.presence_of_all_elements_located(
            ("class name","_aagw")))

            # checking whether number of posts is greater than equal to NumberOfposts user required
            if len(post_elements) >= NumberofPosts:

                # if yes break the loop
                break

            # Else we are updating the curr_height so that we can scroll more and fill more posts
            curr_height = new_height

            # returing number of posts user required
        return post_elements[:NumberofPosts]
        
    except :
        # Handling any encountered errors during this process
        print("Some Error Encountered")
    
```

```python
def likePosts(PostElements):
        
        # Creating a failed_post variable which will store the element that will unable to like
        failed_post =  None
        
        # Iterating through the elements that present in PostElements
        for element in PostElements:
            try:
                # Waiting for the post element to appear 
                time.sleep(1.5)
                
                # Clicking the post
                element.click()
                
                # Again waiting till the window appears with Like button
                time.sleep(1.2)
                
                # Calculating number of likes
                likes = int(wait.until(EC.presence_of_element_located(
                                   ("xpath","//span[contains(@class,'x193iq5w xeuugli x1fj9vlw x13faqbe x1vvkbs xt0psk2 x1i0vuye xvs91rp x1s688f x5n08af x10wh9bi x1wdrske x8viiok x18hxmgj')]")
                    )).text.replace(",","").split()[0])
                
                # locating the like button and clicking it
                Like_btn = wait.until(EC.presence_of_element_located(("class name","_aamw")))
                Like_btn.click()
                
                # Calculating number of likes after Clicking like button
                updated_likes = int(wait.until(EC.presence_of_element_located(
                                   ("xpath","//span[contains(@class,'x193iq5w xeuugli x1fj9vlw x13faqbe x1vvkbs xt0psk2 x1i0vuye xvs91rp x1s688f x5n08af x10wh9bi x1wdrske x8viiok x18hxmgj')]")
                    )).text.replace(",","").split()[0])
                
                # Checking updated_likes is less than likes count 
                if updated_likes < likes:
                    # if yes then post is already liked
                    print("Already liked it")
                    # Again clicking like button to make sure the post is liked
                    Like_btn.click()
                
                # Waiting for the post to get liked
                time.sleep(1.2)
                
                # Going back to "dilsefoodie" Profile
                driver.back()
                
            except:
                # Handling the only one element that is unable to click
                # If triggers then we will store element which is unable to liked by assigning it in failed_post variable
                failed_post = element
                pass
            
        # Checking whether any post is assigned to failed_post or not    
        if failed_post != None:
            failedPostHandler(failed_post,"liked")
```

```python
def UnlikePosts(PostElements):
        
        # Creating a failed_post variable which will store the element that will unable to Unlike
        failed_post =  None
        
        # Iterating through the elements that present in PostElements
        for element in PostElements:
            try:
                # Waiting for the post element to appear 
                time.sleep(1.5)
                
                # Clicking the post
                element.click()
                
                # Again waiting till the window appears with UnLike button
                time.sleep(1.2)
                
                # Calculating number of likes
                likes = int(wait.until(EC.presence_of_element_located(
                                   ("xpath","//span[contains(@class,'x193iq5w xeuugli x1fj9vlw x13faqbe x1vvkbs xt0psk2 x1i0vuye xvs91rp x1s688f x5n08af x10wh9bi x1wdrske x8viiok x18hxmgj')]")
                    )).text.replace(",","").split()[0])
                
                # locating the like button and clicking it
                unLike_btn = wait.until(EC.presence_of_element_located(("class name","_aamw")))
                unLike_btn.click()
                
                # Calculating number of likes after Clicking like button
                updated_likes = int(wait.until(EC.presence_of_element_located(
                                   ("xpath","//span[contains(@class,'x193iq5w xeuugli x1fj9vlw x13faqbe x1vvkbs xt0psk2 x1i0vuye xvs91rp x1s688f x5n08af x10wh9bi x1wdrske x8viiok x18hxmgj')]")
                    )).text.replace(",","").split()[0])
                
                # Checking updated_likes is less than likes count 
                if updated_likes > likes:
                    # if yes then post is already liked
                    print("Already Unliked it")
                    # Again clicking like button to make sure the post is liked
                    Like_btn.click()
                
                
                # Going back to "dilsefoodie" Profile
                driver.back()
                
            except:
                # Handling the only one element that is unable to click
                # If triggers then we will store element which is unable to unliked by assigning it in failed_post variable
                failed_post = element
                pass
        
        # Checking whether any post is assigned to failed_post or not  
        if failed_post!=None:
            failedPostHandler(failed_post,"unliked")
```

```python
def failedPostHandler(failed_post,text):
    try:
            # Waiting for the failed post element to appear 
            time.sleep(1.5)

            # Clicking the failed post
            failed_post.click()

            # Again waiting till the window appears with Like/UnLike button
            time.sleep(1.2)
            
            # Calculating number of likess
            likes = int(wait.until(EC.presence_of_element_located(
                                   ("xpath","//span[contains(@class,'x193iq5w xeuugli x1fj9vlw x13faqbe x1vvkbs xt0psk2 x1i0vuye xvs91rp x1s688f x5n08af x10wh9bi x1wdrske x8viiok x18hxmgj')]")
                    )).text.replace(",","").split()[0])
                
            # locating the like button and clicking it
            like_OR_unLike_btn = wait.until(EC.presence_of_element_located(("class name","_aamw")))
            like_OR_unLike_btn.click()
                
            # Calculating number of likes after Clicking like button
            updated_likes = int(wait.until(EC.presence_of_element_located(
                                   ("xpath","//span[contains(@class,'x193iq5w xeuugli x1fj9vlw x13faqbe x1vvkbs xt0psk2 x1i0vuye xvs91rp x1s688f x5n08af x10wh9bi x1wdrske x8viiok x18hxmgj')]")
                    )).text.replace(",","").split()[0])
             
            # Checking updated_likes is greater than likes count     
            if updated_likes < likes and text == "liked":
                    # if yes then post is already liked
                    print("Already liked it")
                    # Again clicking like button to make sure the post is liked
                    like_OR_unLike_btn.click()
            
            
            # Checking updated_likes is less than likes count 
            elif updated_likes > likes and text=="unliked":
                    # if yes then post is already liked
                    print("Already unliked it")
                    # Again clicking like button to make sure the post is liked
                    like_OR_unLike_btn.click()
                

            # Waiting for the post to get Unliked
            time.sleep(1.2)

            # Going back to "dilsefoodie" Profile
            driver.back()
            
    except:
        print("Some error encountered")
```

```python
# Collecting top 30 posts
postElements = fetchingPosts(30)

# Liking top 30 posts
likePosts(postElements)

# Unliking top 30 posts
UnlikePosts(postElements)
```

### Problem 6 : Extract list of followers
Extract list of followers  
1.Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.  
2.Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

```python
def followersFetcher(text):
    try:
        # Searching and opening the required profile
        SearchOpenProfile(text)

        # Waiting for profile to load properly
        time.sleep(1.5)

        # locating followers button
        followers_btn = wait.until(EC.presence_of_element_located(
        ("xpath","/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[2]/div[2]/section/main/div/header/section/ul/li[2]/a")))

        # Clicking followers button
        followers_btn.click()

        # Waiting for followers window appears
        time.sleep(2)
        
    
        # Collecting all followers insta handles
        followers_names = wait.until(EC.presence_of_all_elements_located(
        ("xpath","//span[contains(@class,'_ap3a _aaco _aacw _aacx _aad7 _aade')]")))

        # Fetching their names
        followers = [name.text for name in followers_names]

        #Locating close button
        close_btn = wait.until(EC.presence_of_element_located(
        ("class name","_abm0"))) 
        
        # Clicking close button
        close_btn.click()

        # returning list of names
        return followers
    
    except:
        # Handling any error that occur while this process
        print("Some Error Encountered")
        
# No one can see or fetch all the followers except ther real user as per new version of instagram
```

```python
def onlyfollowersnotFollowing(followerslist,username):
    for handle in followerslist:
        
        # Searching and opening profile
        SearchOpenProfile(handle)
        
        # Waiting for profile to load properly
        time.sleep(1.5)
        
        # locating following button
        following_btn = wait.until(EC.presence_of_element_located(
        ("xpath","/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[2]/div[2]/section/main/div/header/section/ul/li[3]/a")))
        
        # Clicking following button
        following_btn.click()
        
        # Waiting for following window
        time.sleep(1.5)
        
        # Collecting the instagram handles
        insta_handles = wait.until(EC.presence_of_all_elements_located(
        ("xpath","//div[contains(@class,'x9f619 xjbqb8w x78zum5 x168nmei x13lgxp2 x5pf9jr xo71vjh x1n2onr6 x1plvlek xryxfnj x1c4vz4f x2lah0s x1q0g3np xqjyukv x6s0dn4 x1oa3qoh x1nhvcw1')]")))
        
        # Collecting the insta handles name and converting in set for faster retrieval
        handles_names = set([i.text for i in insta_handles])
        
        if username not in handles_names:
            print(handle)
            
        # Locating close button
        close_btn = wait.until(EC.presence_of_element_located(
            ("class name","_abl-"))) 

        # Clicking close button
        close_btn.click()
        
        # Waiting for closing window properly
        time.sleep(1.8)
```

```python
def findingfollowersnotfollowing(text,username):
    # Searching and opening the required profile
    SearchOpenProfile(text)

    # Waiting for profile to load properly
    time.sleep(1.5)


    # locating followers button
    followers_btn = wait.until(EC.presence_of_element_located(
    ("xpath","/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[2]/div[2]/section/main/div/header/section/ul/li[2]/a")))

    # Clicking followers button
    followers_btn.click()
    
    # Waiting for followers window appears
    time.sleep(2)
    
    # Followers handles with follow/following button of foodtalkindia
    accounts = wait.until(EC.presence_of_all_elements_located(
    ("xpath","//div[contains(@class,'x1dm5mii x16mil14 xiojian x1yutycm x1lliihq x193iq5w xh8yej3')]"))
              )
    
    # Fetching those account details
    accounts = [i.text.split() for i in accounts]
    
    # Storing only those account which I am following
    account = [i[0] for i in accounts if i[-1]=="Following"]
    
    #Locating close button
    close_btn = wait.until(EC.presence_of_element_located(
        ("class name","_abm0"))) 
        
    # Clicking close button
    close_btn.click()
    
    onlyfollowersnotFollowing(account,username)
```

```python
# Fetching account names that i follow from foodtalkindia but they dont follow me
findingfollowersnotfollowing("foodtalkindia",username)
```

### Problem 7 : Check the story of ‘coding.ninjas’
Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -  
1.If You have already seen the story.  
2.Or The user has no story.  
3.Or View the story if not yet seen.  

```python
def seenStory(profilename):
    # Searching and opening the required profile
    SearchOpenProfile(profilename)

    # Waiting for profile to load properly
    time.sleep(1.5)

    # Calculating height before seeing story
    height = wait.until(EC.presence_of_element_located(
     ('class name','_aarh'))).get_attribute('height')
    
    # Checking the height if yes then it has story
    if height == '208':   
        # Saying you havent seen story yet
        print("You haven't seen the story now you will see the story" )
        
        # fetching story button
        story_btn = wait.until(EC.presence_of_element_located(
        ("xpath",'/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[2]/div[2]/section/main/div/header/div/div/span/img'))
                                  )
        # Clicking story button
        story_btn.click()

        # Waiting for story to open
        time.sleep(1.5)

        # Fetching close button
        close_btn = wait.until(EC.presence_of_element_located(
        ('xpath','/html/body/div[2]/div/div/div[3]/div/div/div[3]/div/section/div[3]/div/div'))
                                  )
        # Clicking close button
        close_btn.click()
            
        # Saying now you have seen story successfully
        print('Now you have seen story successfully')
    else:
        # If error occurs that means there is not story
        print("There is no story")
```

```python
# checking story seen or there is no story
seenStory("coding.ninjas")
```
